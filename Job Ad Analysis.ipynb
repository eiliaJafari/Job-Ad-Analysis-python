Pre-processing
</font>
</h2>

<p style="text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
The pre-processing of this project includes three steps, after applying them, the dataset is prepared for analysis.
</font>
</p>


<h3 style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
First step
</font>
</h3>

<p  style="text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
The rows where <b>all 3 columns</b> <code>Level</code>, <code>Salary</code> and <code>Collaboration Type</code> are missing (<code>NaN</code>) are completely removed from the data. With the data obtained from this step, I go to the second step.
</font>
</p>


```python
# Remove rows wihout "Salary", "Level" and "Collaboration Type" 
df.dropna(subset=['Collaboration Type', 'Level', 'Salary'], how='all', inplace=True)
df
```

<h3  style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Second step
</font>
</h3>

<p  style="text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
I replaced the <code>Salary</code> column values with the number they have.
For example,I converted something like <code>>6MT</code> To <code>6</code>.
In general, I separated the number of each string from the <code>Salary</code> column and replace it in the dataset.
    </font>
</p>

```python
# Extract "Salary" values
df['Salary'] = df['Salary'].astype('str').str.extractall('(\d+)').unstack().fillna('').sum(axis=1).astype(int)
df
```

<h3 style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Third step
</font>
</h3>

<p style="text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
I filled in the missing values of the <code>Salary</code> column with average salaries <b>by level</b>. That is, if the <code>Salary</code> of a junior ad (<code>J</code>) was missing, I filled it with the average of all juniors.  
</font>
</p>

```python
# Fill nan values in "Salary" based on "Level"
level_grouped = df.groupby('Level')['Salary'].agg('mean')
mean_I = level_grouped.iloc[0]
mean_J = level_grouped.iloc[1]
mean_M = level_grouped.iloc[2]
mean_S = level_grouped.iloc[3]

df.loc[(df.Salary.isna()) & (df.Level == 'S'), 'Salary'] = mean_S
df.loc[(df.Salary.isna()) & (df.Level == 'M'), 'Salary'] = mean_M
df.loc[(df.Salary.isna()) & (df.Level == 'J'), 'Salary'] = mean_J
df.loc[(df.Salary.isna()) & (df.Level == 'I'), 'Salary'] = mean_I

df
```

<font face="vazir" size=3>
The pre-processing part is finished.</font>

<h3  style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
First step
</font>
</h3>

<p  style="text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
How much has the rate of job ads offering remote work increased compared to before the first official case of Corona in Iran? The meaning of "rate of job ads that offer remote work" is the result of dividing job ads that offer remote work by the total number of ads (the number of dataframe rows). The goal is to find out the impact of Corona on remote work. To solve this problem, I subtracted the rate of remote work ads before Corona from the rate of remote work ads after Corona.
</font>
</p>

```python
start_of_corona = datetime.strptime("2020-02-19", "%Y-%m-%d")
# Find remote_growth
remote_offers_after_corona = len(df.loc[(df['Offers Remote'] == True) & (df['Publish Time'].apply(lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M:%S")) >= start_of_corona)])
remote_offers_before_corona = len(df.loc[(df['Offers Remote'] == True) & (df['Publish Time'].apply(lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M:%S")) < start_of_corona)])
df_len = len(df)

remote_growth =  (remote_offers_after_corona - remote_offers_before_corona) / df_len
remote_growth
```

<h3  style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Second step
</font>
</h3>

<p style="text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
In this step, I intend to discretize time, so I divided the 24 hours into 4 categories according to the table below.
</font>
</p>

<center>
<div  style="line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    
|Category|Range|
|:-------:|:-----:|
|dawn|00:00 to 5:59:59|
|morning|06:00 to 11:59:59|
|noon|12:00 to 17:59:59|
|night|18:00 to 23:59:59|
    
</font>
</div>
</center>

<p style="text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
Now I will calculate how many job advertisements have been published in each of these 4 periods?
</font>
</p>


```python
from datetime import time

dawn = time(hour=6,minute=0,second=0)
morning = time(hour=12,minute=0,second=0)
noon = time(hour=18,minute=0,second=0)
night = time(hour=0,minute=0,second=0)

def discretize_time (time) :
  if time.time() < dawn : 
    return "dawn"
  elif time.time() < morning :
    return "morning"
  elif time.time() < noon :
    return "noon"
  else : return "night"
```

```python
discrete_time = df.groupby(df['Publish Time'].apply(lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M:%S")).apply(discretize_time))['Publish Time'].agg('count').sort_values(ascending=False).rename_axis("index").rename("discrete_time")
```

<p  style="text-align: justify;line-height:200%;font-family:vazir;font-size:medium">

In this step, I want to check the salary of the employees of the data field. In order to be able to filter job ads in this field, I wrote a list of keywords. The title of an ad does not have to be one of the keys in the list of keywords; Just if a key is part of the title of a job ad, I assume that job ad is data-related.

Now I will examine this: What is the average salary of people in the field of data according to their level?   
</font>
</p>


```python
keywords = ['machine learning', 'machinelearning', 'داده' , 'data scientist' ,  'datascientist' ,\
        'هوش مصنوعی' ,'پردازش ویدئو' , 'data engineer' , 'dataengineer' ,'بینایی ماشین' , 'یادگیری ماشین' ,\
        'deep learning', 'deeplearning', 'یادگیری عمیق', 'دیتاساینتیست' , 'artificial intelligence' \
        ,'artificialintelligence', 'هوش' , 'data analysis' , 'dataanalysis' , 'پردازش تصویر' , 'شبکه‌های عمیق', 'علم‌داده']
```
